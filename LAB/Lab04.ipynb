```python
#Exercise 3
#a
print("a) ∃x ∈ Z, 0 <= x <= 100 such that x^3 < x")
for x in range(101):
    if  x**3 < x:
        print("The given statement is Incorrect")
        break
else:
    print("The given statement is Correct")

#b
print("b) ∃x ∈ Z, 0 <= x <= 100 and x is even such that x*3+1 is not a prime number")
def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

for x in range(101):
    if  (is_prime(x*3+1) and x % 2 == 0) :
        print("The given statement is Incorrect")
        break
else:
    print("The given statement is Correct")

#c
print("c) ∃x ∈ Z, 1 <= x <= 100 and x is even such that x*5+3 is not a prime number")

for x in range(100):
    if  (is_prime(x*5+3) and x % 2 == 0) :
        print("The given statement is Incorrect")
        break
else:
    print("The given statement is Correct")

#d
print("d) ∃x ∈ Z, 0 < c <= 100, c % 4 = 0 such that ∀a,b ∈ Z+, c**2 != a**2 + b**2")
for c in range(100):
  if (c % 4 == 0):
    for a in range(1, c):
      for b in range(1, c):
        if (c**2 == a**2 + b**2):
          print("The given statement is Correct")
          break
      else:
        continue
      break
    else:
        continue
    break
else:
    print("The given statement is Incorrect")

#d
print("d) ∃x ∈ Z, 0 < c <= 100, c % 5 = 0 such that ∀a,b ∈ Z+, c**2 != a**2 + b**2")
for c in range(100):
  if (c % 5 == 0):
    for a in range(1, c):
      for b in range(1, c):
        if (c**2 == a**2 + b**2):
          print("The given statement is Correct")
          break
      else:
        continue
      break
    else:
        continue
    break
else:
    print("The given statement is Incorrect")

#f
print("∀c ∈ Z, 0 < c <= 100 such that c**2 > c")
for c in range(100):
  if (c**2 <= c):
    print("The given statement is Correct")
    break
else:
    print("The given statement is Incorrect")
```
